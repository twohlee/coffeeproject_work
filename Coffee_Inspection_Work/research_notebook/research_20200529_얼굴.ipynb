```python
# 필요한 모듈 불러오기
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import torch
from time import sleep
```

```python
img = Image.open('../data/사전테스트데이터/사람1.png')
img = np.array(img)
print(img.shape)
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
```

```python
thres = 170
# ret, binary = cv2.threshold(gray,thres,255, cv2.THRESH_BINARY)
binary = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 101, 5)
# binary = cv2.bitwise_not(binary)
plt.imshow(binary)
```

```python
binary.shape
```

```python
height, width = binary.shape
# 바이너리 이미지와 원본 이미지의 데이터가 일치하는 부분 만 다시 추출히기
tmp = np.zeros_like(img)
for y in range(height):
    for x in range(width):
        # print(y, x)
        if (binary != 0)[y,x]:
            for i in range(4):
                tmp[y,x,i] = img[y,x,i]

src_processed = tmp   
# 이미지 크기 정의
```
