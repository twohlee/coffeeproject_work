# Normal data 백그라운드 -> White

```python
# 이미지 불러오기
angles = [0, 45, 90, 135, 180, 225, 270, 315]
take_list = list()
for take_num in range(14):
    file_list = list()
    for file_num in range(121):
        angle_list = list()
        for angle in angles:
            path = '/Users/gabriel8452/Desktop/coffeebean_project/Coffee_Inspection_preprocessing/data/normal_data/normal_rotated_data/20200521_' + str(file_num)+ '_normal_take' + str(take_num) + '_final_' + str(angle) +'.jpg'
            img = np.array(Image.open(path))
            angle_list.append(img)
        file_list.append(angle_list)
    take_list.append(file_list)
```

```python
# 데이터 처리
angles = [0, 45, 90, 135, 180, 225, 270, 315]
for take_num, file_list in enumerate(take_list):
    for file_num, angle_list in enumerate(file_list):
        for angle_idx, img in enumerate(angle_list):


            # 이미지 배경 White로 바꾸기
            for y in range(64):
                for x in range(64):
                    for z in range(3):
                        if img[y,x,z] <=40:
                            img[y,x,z] = 255

            save_path = '/Users/gabriel8452/Desktop/coffeebean_project/Coffee_Inspection_preprocessing/data/normal_data/normal_third_processed_data/20200525_' + str(file_num) + '_normal_take' + str(take_num)+'_' +  str(angles[angle_idx]) + '_white.jpg'
            Image.fromarray(img).save(save_path)
```
