# 데이터 명암 조절 처리
## 아이디어
    - 데이터 값에 bias 걸어주기
    - 데이터 값에 random noise bias 걸어주기

```python
# 바이어스로 명암조절
tmp1 = np.ones_like(img_array)
tmp2 = img_array * tmp1
bias = 80
for c in range(3):
    for y in range(64):
        for x in range(64):
            if tmp2[y,x,c] != 0 :
                tmp2[y,x,c] = tmp2[y,x,c] + bias
plt.imshow(tmp2)
```

```python
# 인자곱으로 명암 조절하기
tmp1 = np.ones_like(img_array)
tmp2 = img_array * tmp1
alpha = 1.40
for c in range(3):
    for y in range(64):
        for x in range(64):
            if tmp2[y,x,c] != 0 :
                tmp2[y,x,c] = tmp2[y,x,c] * alpha
plt.imshow(tmp2)
```

```python
tmp1 = np.ones_like(img_array)
tmp2 = img_array * tmp1
bias = 20
for c in range(3):
    for y in range(64):
        for x in range(64):
            if tmp2[y,x,c] != 0 :
                tmp2[y,x,c] = tmp2[y,x,c] + (bias * random())
plt.imshow(tmp2)
```
