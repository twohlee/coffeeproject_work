# 데이터 불러오기

```
# 필요한 모듈 불러오기
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```
# 데이터 로드
img = Image.open('/content/drive/My Drive/Coffee_Inspection/참고자료/output/coffee7_contour.png')
img
```

```
img_array = np.array(img)
print(img_array.shape)
```

## 이미지 데이터 처리

```
# 필요한 모듈 불러오기
import torch
```

```
img_tensor = torch.FloatTensor(img_array)
img_tensor.size()
```

```
# 이미지 reshape
img_reshaped = img_tensor.reshape(img_tensor.size()[0] * img_tensor.size()[1]*3)
img_reshaped.size()
```
